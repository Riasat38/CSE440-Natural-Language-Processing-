```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

txt = reuters.raw('training/9865')
#txt = gutenberg.raw('blake-poems.txt')

wrd_token = word_tokenize(txt)

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in wrd_token if word.lower() not in stop_words]

print("Word Tokenization : ",wrd_token)
print("Word Tokens without stopwords: ",filtered_words)
#print(txt)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed_list = []
for word in filtered_words:
  stemmed_list.append(stemmer.stem(word))
print("Stemmed Word:", stemmed_list)

original_freq = FreqDist(wrd_token)
processed_freq = FreqDist(stemmed_list)
print(original_freq.most_common(17))
print(processed_freq.most_common(17))
```

```python
import matplotlib.pyplot as plt

or_wr = [w[0] for w in original_freq.most_common(17)]
or_fr = [w[1] for w in original_freq.most_common(17)]

pr_wr = [w[0] for w in processed_freq.most_common(17)]
pr_fr = [w[1] for w in processed_freq.most_common(17)]

plt.figure(figsize=(12,6))
plt.hist(or_wr, weights = or_fr, alpha=0.7, label='Original')
plt.hist(pr_wr, weights = pr_fr, alpha=0.7, label='Processed')
print(or_wr)
print(pr_wr)

plt.xticks(rotation=45, ha='right')
plt.legend()
plt.title('Top 17 Words Before and After Preprocessing')
plt.show()

```

```python
from nltk.corpus import udhr
import numpy as np

#print(udhr.fileids())
spn = udhr.raw('Spanish_Espanol-Latin1')
#spn = gutenberg.raw('blake-poems.txt')
#print(spn)
def extract_consonants(text):
    wrd_token = word_tokenize(text)
    stop_words = set(stopwords.words('spanish'))
    filtered_words = [word for word in wrd_token if word.lower() not in stop_words]

    pattern = []
    for word in filtered_words:
        i=0
        count = 0
        flag = False
        while i<len(word):
            if word[i].lower() not in 'aeiou':
                flag = True
                count += 1
                if  count == 3:
                    pattern.append(word)
                    break
            else:
                flag = False
                count = 0
            i+=1
    data =  FreqDist(pattern)
    x = [w[0] for w in data.most_common(10)]
    y = [w[1] for w in data.most_common(10)]

    plt.figure(figsize=(12,6))
    plt.hist(x,weights = y, alpha = 0.7)
    plt.xticks(rotation=45, ha='right')
    plt.legend()
    plt.show()

    unique = set(pattern)
    word_to_index = {word: i for i, word in enumerate(unique)}

    matrix = np.zeros((len(unique), len(unique)), dtype=int)

    for i in range(len(pattern) - 1):
        word1, word2 = pattern[i], pattern[i + 1]
        if word1 in word_to_index and word2 in word_to_index:
            matrix[word_to_index[word1]][word_to_index[word2]] += 1
            matrix[word_to_index[word2]][word_to_index[word1]] += 1

    print("Unique words and their indices:")
    for word, index in word_to_index.items():
        print(f"{word}: {index}")

    print("\nNumber of unique words:", len(unique))
    print("Matrix shape:", matrix.shape)

    print("\nCo-occurrence Matrix:")
    print("   ", "  ".join(unique))
    for i, word in enumerate(unique):
        print(f"{word:5} {matrix[i]}")


extract_consonants(spn)
```
